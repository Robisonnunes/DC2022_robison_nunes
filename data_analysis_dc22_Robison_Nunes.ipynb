A Stone tem como objetivo ajudar o empreendedor brasileiro a vender, gerir e crescer

Nosso produto  diferenciado: os clientes pagam seus emprstimos atravs de um

Durante a pandemia, naturalmente tivemos um aumento de clientes inadimplentes, ou

Qual  a curva ideal de vezes que acionamos um cliente?

Quando um cliente comea a apresentar dificuldade na liquidao de seus contratos, ele
passa a ser elegvel para ser acionado por nossa rgua de comunicao com intuito de
estimular a retomada do ritmo saudvel de pagamento de acordo com o perfil e o momento de
cada cliente.

As comunicaes so disparadas por contrato, sendo assim, um cliente pode
receber diversas comunicaes com diferentes contedos, um para cada momento da jornada
de seu respectivo contrato. Desde o envio da comunicao at o recebimento, so gerados flags
que indicam o engajamento do cliente com as comunicaes da seguinte forma:

1. Envio do comunicado > 2. Comunicado recebido > 3.Comunicado lido

Dada a sequncia da comunicao acima, ocasionalmente so observados erros no
processo de envio do comunicado, gerando um percentual (%) de falhas por diversos

Para mensurar a efetividade de uma ao, temos dois timos termmetros: os dias sem

principal. Neste ltimo conceito, mesmo que o contrato apresente algum pagamento, se este montante no for suficiente para cobrir juros + impostos, valores deduzidos prioritariamente,

```python
# Bibliotecas usadas no projeto
import pandas as pd
import functools
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")
```

```python
#functions construidas para esse projeto
def info_data(data):
    
    """ please inform dataset - funo criada para visualizar as principais mtricas dos datasets"""
    
    missing = pd.DataFrame(data = (data.isnull().sum()/len(data))*100, columns=['% missing'])
    unique = pd.DataFrame(data=data.nunique(), columns=['unique'])
    count = pd.DataFrame(data=data.count(), columns=['count'])
    types = pd.DataFrame(data=data.dtypes, columns=['types'])
    dfs = [types,count, missing, unique]
    info = functools.reduce(lambda left,right: pd.merge(left,right,left_index=True, right_index=True), dfs)
    info['duplicated'] = info['count'] - info['unique']
    return info

def varqual(data, col, order, title):    
    """ Plotar um grfico de freq. (argumentos: dataset, feature, title, ordenamento)"""
    print(col)
    print('*' * 80, '\n')
    print(((data[col].value_counts(normalize=True)*100)).round(2))

    fig, (ax1) = plt.subplots(1,1 ,figsize=(18,5))
    ax = sns.countplot(data = data, x = col, order=order,palette='crest')
    ax.set_title(title)
    
    for i in ax.patches:
        ax.annotate(i.get_height(),
                    (i.get_x() + i.get_width() / 2, i.get_height()),
                    ha='center',
                    va='baseline',
                    fontsize=12,
                    color='black',
                    xytext=(0, 1),
                    textcoords='offset points')    
    plt.show() 

def qual_qual(data, col, col2 ):
    """ Plota grfico de barras (argumentos: dataset, feature, feature_target)"""
    analise_qual = data.groupby([col, col2]).size().unstack()
    ax = (analise_qual.T*100/analise_qual.T.sum()).T.plot(kind='bar',
    figsize=(15,6),width=0.8, color=['gray', 'green','greenyellow','red'],stacked=True, rot=0 )
    ax.set_title('Relaao entre ' + col + ' x ' + col2, fontsize=15)
    
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x ,y = p.get_xy()
        ax.annotate('{:.2f}%'.format(height), (x + width/2, y + height*0.2),
                    ha='center',
                    fontsize=15,
                    color='black',
                    xytext=(0, 1),
                    textcoords='offset points')
    
    plt.legend(fontsize=12, 
    bbox_to_anchor = (1.05, 0.6))
    
    plt.show()    
```

### <font color=green>  3.1 -  PORTIFLIO COMUNICADOS</font>

```python
# #ELT - COMUNICADOS ==================================================================================

# Salvando em novo objeto para manter o histrico e viabilizar um row back
p_comunicados1 = p_comunicados.copy()

# Criando chave nica para merge na granularidade de data relacionamento de um contrato
#dentro da tabela portiflio geral
p_comunicados1['pk_comunicados'] = p_comunicados1['contrato_id'] + '-'+ p_comunicados1['dt_ref_portfolio']

# Ordenando os dados
p_comunicados1 = p_comunicados1.sort_values(by=['pk_comunicados', 'tipo_acao'])

# reconstruindo o index
p_comunicados1 = p_comunicados1.reset_index(drop=True)

# For para agrupar em uma linha o tipo de acao e flag de engajamento da mesma na granularidade de data
p_comunicados1['tp_acao1'] = ''
p_comunicados1['st_acao1'] = ''
p_comunicados1['tp_acao2'] = ''
p_comunicados1['st_acao2'] = ''

for i in range(len(p_comunicados1)-1):
    if p_comunicados1['pk_comunicados'][i] == p_comunicados1['pk_comunicados'][i+1]:
        p_comunicados1['tp_acao1'][i] =  p_comunicados1['tipo_acao'][i]
        p_comunicados1['st_acao1'][i] =  p_comunicados1['status'][i]
        p_comunicados1['tp_acao2'][i] =  p_comunicados1['tipo_acao'][i+1]
        p_comunicados1['st_acao2'][i] =  p_comunicados1['status'][i+1]
        i += 2
    else:
        if p_comunicados1['tp_acao1'][i] == 'EMAIL':
            p_comunicados1['tp_acao1'][i] =  p_comunicados1['tipo_acao'][i] 
            p_comunicados1['st_acao1'][i] =  p_comunicados1['status'][i]
        else:        
            p_comunicados1['tp_acao2'][i] =  p_comunicados1['tipo_acao'][i]
            p_comunicados1['st_acao2'][i] =  p_comunicados1['status'][i]
        i += 1

# Eliminando as linhas duplicadas que sobraram aps o for acima
p_comunicados1 = p_comunicados1.drop_duplicates(subset='pk_comunicados', keep='first')

# reconstruindo o index
p_comunicados1 = p_comunicados1.reset_index(drop=True)

# Elimando colunas redundantes
p_comunicados1.drop(columns=['tipo_acao','status'], inplace=True)
```

```python
# Criando a feature engajamento 
p_comunicados1['engajamento'] = ''

for i in range(len(p_comunicados1['contrato_id'])):
    if ((p_comunicados1['st_acao1'][i] == 'LIDO') | (p_comunicados1['st_acao1'][i] == 'RESPONDIDO')) | ((p_comunicados1['st_acao2'][i] == 'LIDO') | (p_comunicados1['st_acao2'][i] == 'RESPONDIDO')):
        p_comunicados1['engajamento'][i] = 1
    else:
        p_comunicados1['engajamento'][i] = 0

# salvando o output do etl
p_comunicados1.to_csv('dados/p_comunicados.csv', sep=',', encoding='utf-8', index=False)
```

### <font color=green>  3.2 -  PORTIFLIO TPV</font>

```python
#ELT - TPV ========================================================================================

# Salvando em novo objeto para manter o histrico e viabilizar um row back
p_tpv1 = p_tpv.copy()

#Corrigindo o formato da data
p_tpv1['dt_transacao'] = pd.to_datetime(p_tpv1['dt_transacao'], format='%Y%m%d', errors='ignore')
p_tpv1['safra_transacao'] =  p_tpv1['dt_transacao'].apply(lambda x: x.strftime('%b-%Y')) 
p_tpv1['dt_transacao'] = p_tpv1['dt_transacao'].astype(str)

# Criando chave unica
p_tpv1['pk_tpv'] = p_tpv1['nr_documento'] + '-'+ p_tpv1['dt_transacao']
p_tpv1.sort_values(by='dt_transacao', ascending=True, inplace=True)
# salvando o output do etl
p_tpv1.to_csv('dados/p_tpv.csv', sep=',', encoding='utf-8', index=False)
```

### <font color=green>  3.3 -  PORTIFLIO CLIENTES</font>

```python
#ELT - PORTIFLIO CLIENTES==========================================================================

# Salvando em novo objeto para manter o histrico e viabilizar um row back
p_clientes1 = p_clientes.copy()

# Excluindo as linhas duplicadas
p_clientes1.drop_duplicates(subset='nr_documento', keep='first', inplace=True)

# Excluindo a incosistencia do estado 'ND'
p_clientes1 = p_clientes1[p_clientes1['estado']!= 'ND']

#  salvando o output do etl
p_clientes1.to_csv('dados/p_clientes.csv', sep=',', encoding='utf-8', index=False)
```

### <font color=green>  3.4 -  CRIACAO DA BASE NICA</font>

```python
#ELT - BASE NICA ==================================================================================
# Load dos datasets transformados pelas etapas de ETL acima
p_comunicados1 =  pd.read_csv('dados/p_comunicados.csv', sep=',', encoding='utf-8') 
p_tpv1 =  pd.read_csv('dados/p_tpv.csv', sep=',', encoding='utf-8')
p_clientes1 =  pd.read_csv('dados/p_clientes.csv', sep=',', encoding='utf-8')

# Criao das chaves estrangeiras no portiflio geral
p_geral['pk_comunicados'] = p_geral['contrato_id'] + '-'+ p_geral['dt_ref_portfolio']
p_geral['pk_tpv'] = p_geral['nr_documento'] + '-'+ p_geral['dt_ref_portfolio']

# Join portifolio geral x portifolio comunicados atribuindo ao objeto df
df = p_geral.merge(p_comunicados1, how='left', on='pk_comunicados')

# Join do objeto df x portifolio tpv
df = df.merge(p_tpv1, how='left', on='pk_tpv')

# Corrigindo a chave estrangeira no objeto df para o join com o portfolio clientes
df.rename(columns={'nr_documento_x':'nr_documento'}, inplace=True)

# Join do objeto df x portifolio clientes
df = df.merge(p_clientes1, how='left', on='nr_documento')

# Corrigindo o nome de colunas e eliminando as colunas desnecessrias
df.rename(columns={'contrato_id_x':'contrato_id','dt_ref_portfolio_x':'dt_ref_portfolio'}, inplace=True)
df.drop(columns=['pk_comunicados','pk_tpv','contrato_id_y','dt_ref_portfolio_y','nr_documento_y'], inplace=True)
```

#### <font color=green>  4.2.1 - Grfico de Elbon - Kmeans</font>

```python
# Ser usado o algoritmo de machine learning Kmeans para clusterizao da base

# Lista do WCSS (within-cluster sum of squares)
wcss = []

# Grdico de Elbom ele mostra o n de cluster ideal (hiperparemetro do modelo)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", n_init=10, max_iter=300)
    X1 = cluster.iloc[ :, 1:].values
    kmeans.fit_predict(X1)
    wcss.append(kmeans.inertia_) 
        
plt.figure(1 , figsize = (16 ,5))    
plt.plot( wcss, 'ro-', label="WCSS", color='green')
plt.title("Computing WCSS for KMeans++")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# O que esse loop faz?
# Ele cria um nova feature no df0 referente a nova regua de cobrana atribuindo a acao ano macardor 
# de dspp e dsp

df1['acao_1'] = ''

for i in range(len(df1['contrato_id'])):
    
    if df1['dsp'][i]==2:
        df1['acao_1'][i] = 'dsp1_campanhaobservacao'
    
    elif df1['dsp'][i]==7:
        df1['acao_1'][i] = 'dsp2_campanhaparcelamento'
    
    elif df1['dsp'][i]==12:
        df1['acao_1'][i] = 'dsp3_campanhaboletoquitado12'
    
    elif df1['dsp'][i]==20:
        df1['acao_1'][i] = 'dsp4_campanhaprenegativacao'    
    
    elif df1['dsp'][i]==30:
        df1['acao_1'][i] = 'dsp5_campanhanegativacao'
        
    elif df1['dsp'][i]==35:
        df1['acao_1'][i] = 'dsp6_campanhaboletoquitado35' 
        
    elif df1['dspp'][i]==3:
        df1['acao_1'][i] = 'dspp1_campanhaobservacao'
    
    elif df1['dspp'][i]==8:
        df1['acao_1'][i] = 'dspp2_campanhaparcelamento'
    
    elif df1['dspp'][i]==13:
        df1['acao_1'][i] = 'dspp3_campanhaboletoquitado' 

# salvando o novo df1 com a nova marcaao de acoa da regua escolhida
df1.to_csv('dados/df_cluster1.csv', sep=',', encoding='utf-8', index=False)        
```

```python
# Fazendo o load do df1
df1 = pd.read_csv('dados/df_cluster1.csv', sep=',', encoding='utf-8')

# Criando o dataframe para o grfico para simular a aplicaao da regua de cobrana
df_1 = df1.copy()
df_1 = df_1[df_1['acao_1']!='']
df_1 = df_1.groupby(by=['contrato_id','acao_1'])['acao_1'].count()
df_1 = pd.DataFrame(df_1)
df_1.rename(columns={'acao_1': 'acionamentos_1'}, inplace=True)

df_1 = df_1.reset_index()

df_1['efetividade_1'] = ""
df_1['tx_efetividade_1'] = ""

df_1 = df_1[['contrato_id','acao_1','acionamentos_1','efetividade_1','tx_efetividade_1']] 

# Qual  a lgica aplicada aqui?
# Para cada linha indenfifica-se qual  a campanha aplicada, depois na coluna 'efetividade'  alocado o 
# resultado da multiplciaao da taxa de efetividade da acao identificada  do cluster 1 (p_campanha1)
# com a soma dos acionamentos para campanha identificada, e na coluna 'tx_efetividade'  o resultado da 
# coluna efetividade dividida pelo n de acionamentos (groupby por contrato_id e acao_1)

for i in range(len(df_1)):
    if df_1['acao_1'][i] ==  'dsp1_campanhaobservacao':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.176945
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dsp2_campanhaparcelamento':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.076953
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dsp3_campanhaboletoquitado12':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.031133
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dsp4_campanhaprenegativacao':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.039062
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dsp5_campanhanegativacao':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.012442
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dsp6_campanhaboletoquitado35':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.007505
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dspp1_campanhaobservacao':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.153061
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
            
    elif df_1['acao_1'][i] == 'dspp2_campanhaparcelamento':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.141361
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]
    
    elif df_1['acao_1'][i] == 'dspp3_campanhaboletoquitado':
            df_1['efetividade_1'][i] = df_1['acionamentos_1'][i] * 0.101103
            df_1['tx_efetividade_1'][i] = df_1['efetividade_1'][i] / df_1['acionamentos_1'][i]

# aqui  criado a base de dados final para plotar o resultado da simulacao da nova rgua de cobrana
#  a junao de 3 objetos resultado no df_1                           
                
df_1['acionamentos_1'] = df_1['acionamentos_1'].astype(float)            
df_1['tx_efetividade_1'] = df_1['tx_efetividade_1'].astype(float)
                               
df11 = df_1.groupby(by=['contrato_id'])['acionamentos_1'].sum()
df11 = pd.DataFrame(df11) 

df12 = df_1.groupby(by=['contrato_id'])['tx_efetividade_1'].mean()
df12 = pd.DataFrame(df12) 

df_1 = df11.merge(df12, how='left', on='contrato_id')
df_1['contrato_id'] = df_1.index
df_1.reset_index(drop=True, inplace=True)
df_1 = df_1[['contrato_id','acionamentos_1','tx_efetividade_1']]

# Salvando o df_0 na pasta dados
df_1.to_csv('dados/df_curva_cluster1.csv', sep=',', encoding='utf-8', index=False) 
```
